## Exercise 3: Create a model
Analyze the following code. Then, run it and answer the questions.

```python
import warnings
warnings.filterwarnings('ignore')

# Create the linear regression object
lr = lm.LinearRegression()

x = df.Hours_Studied.values.reshape(-1, 1) # Independent variable
y = df.Test_Grade.values                   # Dependent variable

# Train the model using the training sets
lr.fit(x, y)

print('Intercept (c): ', lr.intercept_)
print('Coefficient (m): ', lr.coef_)
```

## Exercise 4: Evaluate a model
Analyze the following code. Then, run it and answer the questions.

```python
# Add predicted value to the data frame
df['Test_Grade_Pred'] = lr.predict(x)

# Calculating R-squared manually
df['SST'] = np.square(df['Test_Grade'] - df['Test_Grade'].mean())
df['SSR'] = np.square(df['Test_Grade_Pred'] - df['Test_Grade'].mean())

print('Sum of SSR:', df['SSR'].sum())
print('Sum of SST:', df['SST'].sum())
print()

print(df)
df.to_csv('r-squared.csv', index = False)

print('\nR-squared using manual calculation:', df['SSR'].sum() / df['SST'].sum())

# The use of the built-in function
print('R-squared using built-in function:', r2_score(df.Test_Grade,  df.Test_Grade_Pred))
print('Mean Absolute Error:', mean_absolute_error(df.Test_Grade, df.Test_Grade_Pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(df.Test_Grade, df.Test_Grade_Pred)))
```

**Questions:**
1. How do you interpret the model evaluation metrics used?
2. How much can a built model be mistaken in predicting the test grade? How will you determine this accuracy?

**Answers:**


1.  
    *   **R-squared (R)**: This metric tells us the proportion of the variance in the dependent variable (Test Grade) that is predictable from the independent variable (Hours Studied). An R-squared of 0.976 (or 97.6%) means that approximately 97.6% of the variation in test grades can be explained by the linear relationship with hours studied. A value close to 1 indicates a very good fit of the model to the data.
    *   **Mean Absolute Error (MAE)**: This is the average of the absolute differences between the actual test grades and the predicted test grades. An MAE of approximately 1.62 means that, on average, the model's predictions are off by about 1.62 points. It gives a straightforward measure of the average prediction error.
    *   **Root Mean Squared Error (RMSE)**: This is the square root of the average of the squared differences between the actual test grades and the predicted test grades. An RMSE of approximately 2.04 means that the typical error in the model's predictions is about 2.04 points. RMSE penalizes larger errors more heavily than MAE, making it sensitive to outliers.

2.  To determine how much the built model can be mistaken in predicting the test grade and quantify its accuracy, we would look primarily at the **Mean Absolute Error (MAE)** and **Root Mean Squared Error (RMSE)**.
    *   The **MAE (1.62)** tells you that, on average, the model's prediction will be about 1.62 points away from the actual test grade.
    *   The **RMSE (2.04)** gives you a sense of the typical magnitude of the prediction errors, considering that larger errors have a greater impact due to the squaring.
    Both metrics provide an estimate of the prediction accuracy, with lower values indicating a more accurate model. In this case, an MAE of 1.62 and an RMSE of 2.04 suggest that the model is relatively accurate in predicting test grades based on the hours studied, with typical errors in the range of 1.6 to 2.0 points.

## Exercise 5: Outliers
Now, we'll introduce an outlier i.e., a student has studied 5 hours and scored 100. Assume that this student has higher IQ than others in this group. Notice the drop in R-squared value. So it is important to apply business logic to avoid including outliers in the training dataset to generalize the model and increase accuracy. Analyze the following code. Then, run it and answer the questions.

```python
# Load data
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Lab04/Grade_Set_1.csv')

df.loc[3] = np.array([5, 100])

x = df.Hours_Studied.values.reshape(-1, 1) # Independent variable
y = df.Test_Grade.values            # Dependent variable

# Train the model using the training sets
lr.fit(x, y)
print('Intercept (c):', lr.intercept_)
print('Coefficient (m):', lr.coef_)
```

```python
# Manual prediction for a given value of x
X = 6 # Hours Studied
print('Manual prediction:', c + m * X) # Insert appropriate values for c and m

# Predict using the built-in predict() function
print('Using predict function:', lr.predict([[6]]))

# Plot the fitted line
plt.scatter(x, y, color = 'black')
plt.plot(x, lr.predict(x), color = 'blue', linewidth = 3)
plt.title('Grade vs Hours Studied')
plt.ylabel('Test_Grade')
plt.xlabel('Hours_Studied')
plt.show()

# Add predicted value to the data frame
df['Test_Grade_Pred'] = lr.predict(x)

# The use of the built-in function
print('R-squared:', r2_score(df.Test_Grade,  df.Test_Grade_Pred))
print('Mean Absolute Error:', mean_absolute_error(df.Test_Grade, df.Test_Grade_Pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(df.Test_Grade, df.Test_Grade_Pred)))
```

**Questions:**
1. How have the model evaluation metrics changed when an outlier is added to the data set?
2. How should outlier values be handled and why?

**Answers:**


1.  When the outlier was added, the model evaluation metrics changed significantly. The **R-squared value dropped** from approximately 0.976 to 0.681. This indicates that the model now explains a much lower percentage of the variance in test grades (only about 68.1%) compared to the original data set. The **Mean Absolute Error (MAE) increased** from about 1.62 to 4.93, meaning the average prediction error is now much larger. Similarly, the **Root Mean Squared Error (RMSE) increased** from about 2.04 to 8.16, showing a substantial increase in the typical magnitude of prediction errors. All these changes indicate that the model's performance has worsened considerably with the inclusion of the outlier.

2.  Outlier values should generally be **carefully considered and potentially handled** (e.g., removed, transformed, or analyzed separately) because they can disproportionately influence the model training process and lead to a model that does not generalize well to the majority of the data. In this specific case, the outlier (a student with low study hours but a very high test grade) pulled the regression line upwards, making the model less accurate for the other data points. Handling outliers, often after applying business logic to understand their cause, helps to build a more robust and accurate model that better represents the typical relationship between the variables.

# Exercise 9: Evaluate a model
We will now evaluate the model for different degrees of polynomial used for modelling the relationship between the dependent and independent variable. In order to simplify calculations, we will use the NumPy's **vander** function that returns powers of the input vector. Analyze the following code. Then, run it and answer the questions.

```python
x = df.Hours_Studied # Independent variable

for deg in range(1, 6):
    lr.fit(np.vander(x, deg + 1), y);
    y_lr = lr.predict(np.vander(x, deg + 1))
    plt.plot(x, y_lr, label = 'degree ' + str(deg));
    plt.legend(loc = 2);
    print('R-squared for degree ' + str(deg) + ' =',  r2_score(y, y_lr))
plt.plot(x, y, 'ok')
plt.show()
```

When examining the R-squared values, the increase from degree 2 (0.961) to degree 3 (0.994) is substantial, indicating that a cubic polynomial captures the relationship much better than a linear or quadratic one. While higher degrees (4 and 5) show slight improvements in R-squared, the gains are marginal (0.9955 and 0.9956 respectively). Choosing a degree of **3 or 4** seems optimal. ## Exercise 10: Create an optimal model
Build the optimal model using the polynomial degree you selected in exercise 9. To do this, set the appropriate value for the **degree** variable. Analyze the following code. Then, run it and answer the questions.

In order to simplify calculations, we will use the **PolynomialFeatures()** function that generates a new feature matrix consisting of all polynomial combinations of the features with degree less than or equal to the specified degree. For example, if an input sample is two dimensional and of the form **[a, b]**, the **degree-2** polynomial features are **[1, a, b, \\(a^2\\), ab, \\(b^2\\)]**.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

x = df.Hours_Studied.values.reshape(-1, 1) # Independent variable
y = df.Test_Grade                   # Dependent variable

degree = 4 # Set the appropriate value
model = make_pipeline(PolynomialFeatures(degree), lr)

model.fit(x, y)

plt.scatter(x, y, color = 'black')
plt.plot(x, model.predict(x), color = 'green')
plt.title('Grade vs Hours Studied')
plt.ylabel('Test_Grade')
plt.xlabel('Hours_Studied')
plt.show()

print('R-squared using built-in function:', r2_score(y, model.predict(x)))
prediction = model.predict([[6]])
print("Test Grade Prediction for 6 hours of study:", prediction)
```
